```python
import json
import requests
import pandas as pd
```

```python
def get_df(start_number):
    test = requests.get('https://www.bitmex.com/api/v1/trade/bucketed?binSize=5m&partial=false&symbol=XBT&count=750&start={}&reverse=true&endTime=2018-11-05T18%3A00%3A00.000Z'.format(start_number))
    a = json.loads(test.text)
    df = pd.DataFrame.from_dict(a)
    
    return df
```

```python
final_df = pd.DataFrame()

for x in list(range(0,8250,750)):
    print(x)
    temp_df = get_df(x)
    final_df = final_df.append(temp_df)
```

```python
final_df = final_df[['timestamp','open','high','low','close','volume']]
```

```python
final_df = final_df.sort_values('timestamp').reset_index(drop=True)
```
